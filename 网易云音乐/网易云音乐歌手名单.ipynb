## 爬取全部歌手名单

参考：https://blog.csdn.net/wanhaiwei/article/details/84327561

```python
import requests
import re
import csv
import json
import time
import pandas as pd
import random

pd.set_option('display.max_rows', None)
```

### 全部歌手

```python
data = pd.DataFrame()

languages = ['华语','欧美','日本','韩国','其他']
types = ['男','女','组合/乐队']
param1 = [1001, 1002, 1003, 2001, 2002, 2003, 6001, 6002, 6003, 7001, 7002, 7003, 4001, 4002, 4003] # 按语言分
param2 = [65, 66, 67, 68, 69, 70, 71, 72, 73, 74, 75, 76, 77, 78, 79, 80, 81, 82, 83, 84, 85, 86, 87, 88, 89, 90,0] # 按字母分
#param1,param2 = [1001, 1002,1003,2001],[65,66]

cnt = 0
for i in param1:
    for j in param2:
        url = 'http://music.163.com/discover/artist/cat?id=' + str(i) + '&initial=' + str(j)
        res = requests.get(url)
        tags = re.findall(r'<a href=".*?/artist\?id=(\d+)" class="nm nm-icn f-thide s-fc0" title=".*?的音乐">(.*?)</a>', res.text, re.S)
        tmp = pd.DataFrame(tags,columns=['Id','Name'])
        tmp['languages'] = languages[cnt//3] # 012华语345欧美678日本91011韩国121314其他
        tmp['types'] = types[cnt%3] # 0 男 1 女 2 组合
        data = data.append(tmp,ignore_index=True)
        print('data.shape:',data.shape)
        time.sleep(random.randint(0,2))
    cnt += 1
```
